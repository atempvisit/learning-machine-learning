```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('titanic_data_updated.csv')


df
```

```python
df.drop(['PassengerId', 'Name', 'Ticket'], axis=1, inplace=True)

df
```

```python
df.isnull().sum()
```

```python
df.drop(['Cabin'], axis=1, inplace=True)

df
```

```python
X = df.drop(['Survived'], axis=1)
X
```

```python
Y = df['Survived']

Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
Y_train
```

```python
X_test
```

```python
X_train.isnull().sum()
```

```python
X_test.isnull().sum()
```

```python
age_mean = X_train['Age'].mean()

X_train['Age Mean Imputed'] = X_train['Age'].fillna(age_mean)

X_test['Age Mean Imputed'] = X_test['Age'].fillna(age_mean)

X_train.isnull().sum()
```

```python
age_median = X_train['Age'].median()

X_train['Age Median Imputed'] = X_train['Age'].fillna(age_median)

X_test['Age Median Imputed'] = X_train['Age'].fillna(age_median)

X_train.isnull().sum()
```

```python
X_train
```

```python
sns.kdeplot(data=X_train, x = 'Age Mean Imputed')
```

```python
sns.kdeplot(data=X_train, x = 'Age Median Imputed')
```

```python
from sklearn.impute import SimpleImputer

age_imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

age_imputer.fit(X_train[['Age']])

X_train['Age'] = age_imputer.transform(X_train[['Age']])

X_train.isnull().sum()

X_train.drop(['Age Mean Imputed', 'Age Median Imputed'], axis=1, inplace=True)

X_train
```

```python
X_test['Age'] = age_imputer.transform(X_test[['Age']])

X_test.drop(['Age Mean Imputed', 'Age Median Imputed'], axis=1, inplace=True)

X_test.isnull().sum()
```

```python
df3 = pd.read_csv('titanic_data_updated.csv')

df3.drop(['PassengerId', 'Name', 'Ticket'], axis=1, inplace=True)

df3.isnull().sum()
```

```python
X3 = df3.drop(['Survived'], axis=1)

Y3 = df3['Survived']

X3_train, X3_test, Y3_train, Y3_test = train_test_split(X3, Y3, test_size=0.2, random_state=42)
```

```python
age_imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

age_imputer.fit(X3_train[['Age']])

X3_train['Age'] = age_imputer.transform(X3_train[['Age']])
X3_test['Age'] = age_imputer.transform(X3_test[['Age']])
```

```python
embarked_imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')

embarked_imputer.fit(X3_train[['Embarked']])

X3_train['Embarked'] = embarked_imputer.transform(X3_train[['Embarked']]).ravel()

X3_test['Embarked'] = embarked_imputer.transform(X3_test[['Embarked']]).ravel()


X3_train.isnull().sum()
X3_test.isnull().sum()
```

```python
sns.countplot(data=X3_train, x='Cabin')
```

```python
cabin_imputer = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value='Missing', add_indicator=True)

cabin_imputer.fit(X3_train[['Cabin']])

X3_train[['Cabin', 'cabin_missing_indicator']] = cabin_imputer.transform(X3_train[['Cabin']])

X3_test[['Cabin', 'cabin_missing_indicator']] = cabin_imputer.transform(X3_test[['Cabin']])
```
