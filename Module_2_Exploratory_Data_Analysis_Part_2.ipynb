```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic_data_updated.csv')
```

```python
sns.countplot(x=df['Sex'], hue=df['Survived'])
```

```python
gender_split = df.groupby('Sex')['Survived'].value_counts(normalize=True)

print(gender_split)
```

```python
sns.countplot(x=df['Pclass'], hue=df['Survived'])
```

```python
gender_split = df.groupby('Pclass')['Survived'].value_counts(normalize=True)

print(gender_split)
```

```python
sns.barplot(x=df['Pclass'], y=df['Fare'])
```

```python
sns.barplot(x=df['Survived'], y=df['Age'], hue=df['Sex'])
```

```python
sns.kdeplot(data=df, x='Age', hue='Survived', fill=True)
```
