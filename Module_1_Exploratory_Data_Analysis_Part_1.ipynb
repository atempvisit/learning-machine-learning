```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic_data_updated.csv')

df.shape

print(f'Total potentital features : {df.shape[1]}')
print(f'Total sample data : {df.shape[0]}')
```

```python
sns.countplot(data=df, x=df['Survived'])
```

```python
df['Survived'].value_counts() / len(df) * 100
```

```python
survived_count = df['Survived'].value_counts()

survived_label = df['Survived'].unique()

plt.pie(survived_count, autopct='%1.1f', labels=survived_label)
plt.show()
```

```python
sns.countplot(data=df, x=df['Pclass'])
```

```python
df['Pclass'].value_counts() / len(df) * 100
```

```python
sns.countplot(data=df, x=df['Sex'])
```

```python
df['Sex'].value_counts() / len(df) * 100
```

```python
sns.histplot(data=df, x=df['Age'],bins=50)
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
sns.kdeplot(data=df, x=df['Age'])
plt.xlabel('Age')
plt.ylabel('Probability Density')
plt.show()
```

```python
sns.histplot(data=df, x=df['Fare'],bins=50)
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```

```python
sns.kdeplot(data=df, x=df['Fare'])
plt.xlabel('Fare')
plt.ylabel('Probability Density')
plt.show()
```

```python
sns.boxplot(df['Fare'])

plt.show()
```
